# Representing Systems of Equations as Matrices #

```python
import numpy as np 
import matplotlib.pyplot as plt
```

# Linear algebra #

$$
\begin{cases} 
-x_1+3x_2=7, \\ 3x_1+2x_2=1, \end{cases}\tag{1}
$$

```python
equation = np.array([
    [-1, 3],
    [3, 2]
],dtype=np.dtype(float))

result = np.array([7, 1], dtype=np.dtype(float))
print(f"equation is {equation}")
print(f"The result is {result}")
```

```python
# Linear algebra function to solve problems

x = np.linalg.solve(equation, result)
print(f"Solution are => {x}")
```

```python
print(f"So : \nx1 = {x[0]} \nx2 = {x[1]}")
```

```python
determinant = np.linalg.det(equation)
print(f"Matrix determinant => {determinant:.1f} ")
```

```python
# Ploting
system = np.hstack((equation,result.reshape((2,1))))
print(f"system => {system}")
```

```python
plt.plot(equation, result)
plt.title("System ploting")
plt.show()
```
